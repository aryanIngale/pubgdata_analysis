### Step 2 : Preforming EDA

```python
#reading/loading datasets

values = pd.read_csv("train_values.csv")
labels = pd.read_csv("train_labels.csv")
df = values.merge(labels, on="building_id", how="inner")
df.head()
```

```python
cat_cols = df.select_dtypes(include=["object"]).columns.tolist()
# Checking a few categorical distributions
n_show = min(6, len(cat_cols))
fig, axes = plt.subplots(nrows=n_show, ncols=1, figsize=(10, 3*n_show))
if n_show == 1:
    axes = [axes]
for ax, col in zip(axes, cat_cols[:n_show]):
    df[col].value_counts().head(10).plot(kind="bar", ax=ax)
    ax.set_title(f"{col} (top 10)")
    ax.set_xlabel("")
plt.tight_layout()
plt.show()
```

```python
# Copy to avoid modifying original
data = df.copy()

#  handling missing values
for col in data.columns:
    if data[col].dtype == "object":
        data[col] = data[col].fillna("missing")
    else:
        data[col] = data[col].fillna(data[col].median())

#cap outliers on a few skewed numeric columns
def cap_outliers(series, k=1.5):
    q1, q3 = series.quantile(0.25), series.quantile(0.75)
    iqr = q3 - q1
    lower, upper = q1 - k*iqr, q3 + k*iqr
    return np.clip(series, lower, upper)

maybe_cap = ["age", "area_percentage", "height_percentage", "count_floors_pre_eq"]
for c in maybe_cap:
    if c in data.columns and np.issubdtype(data[c].dtype, np.number):
        data[c] = cap_outliers(data[c])

print("Cleaning done. Nulls left:", data.isnull().sum().sum())
```

```python
from sklearn.preprocessing import LabelEncoder
X = data.drop(columns=["building_id", "damage_grade"])
y = data["damage_grade"] - 1

# Label-encode each categorical column
cat_cols = X.select_dtypes(include=["object"]).columns.tolist()
encoders = {}
for col in cat_cols:
    le = LabelEncoder()
    X[col] = le.fit_transform(X[col].astype(str))
    encoders[col] = le

print("Encoded categorical columns:", len(cat_cols))
X.head()
```

### Step 4 : Model training And evaluting

```python
#Splitting Test-train data

from sklearn.model_selection import train_test_split, StratifiedKFold, cross_val_score
from sklearn.metrics import (accuracy_score, f1_score, classification_report, confusion_matrix)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=42)
X_train.shape, X_test.shape
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
import xgboost as xgb
import lightgbm as lgb

#logistic Regression
logreg = LogisticRegression(max_iter=200, n_jobs=-1, class_weight="balanced", random_state=42)
logreg.fit(X_train, y_train)
y_pred = logreg.predict(X_test) + 1
print("\n=== Logistic Regression ===")
print("Accuracy:", accuracy_score(y_test + 1, y_pred))
print(classification_report(y_test + 1, y_pred))

#random forestclassifier
rf = RandomForestClassifier(
    n_estimators=400, max_depth=None, min_samples_leaf=2,
    n_jobs=-1, random_state=42
)
rf.fit(X_train, y_train)
y_pred = rf.predict(X_test) + 1
print("\n=== Random Forest ===")
print("Accuracy:", accuracy_score(y_test + 1, y_pred))
print(classification_report(y_test + 1, y_pred))

#XG Boost
xgb_model = xgb.XGBClassifier(
    objective="multi:softmax",
    num_class=3,
    n_estimators=600,
    learning_rate=0.05,
    max_depth=8,
    subsample=0.9,
    colsample_bytree=0.9,
    random_state=42,
    n_jobs=-1
)
xgb_model.fit(X_train, y_train)
y_pred = xgb_model.predict(X_test) + 1
print("\n=== XGBoost ===")
print("Accuracy:", accuracy_score(y_test + 1, y_pred))
print(classification_report(y_test + 1, y_pred))

#Lgbm Classifier 
lgbm = lgb.LGBMClassifier(
    objective="multiclass",
    num_class=3,
    n_estimators=700,
    learning_rate=0.05,
    num_leaves=63,
    min_child_samples=40,
    subsample=0.85,
    colsample_bytree=0.85,
    random_state=42
)
lgbm.fit(X_train, y_train)
y_pred = lgbm.predict(X_test) + 1
print("\n=== LightGBM ===")
print("Accuracy:", accuracy_score(y_test + 1, y_pred))
print(classification_report(y_test + 1, y_pred))
```

### Step 5 : picking best Model

```python
results = {}

# Logistic Regression
y_pred = logreg.predict(X_test) + 1
results["Logistic Regression"] = accuracy_score(y_test + 1, y_pred)

# Random Forest
y_pred = rf.predict(X_test) + 1
results["Random Forest"] = accuracy_score(y_test + 1, y_pred)

# XGBoost
y_pred = xgb_model.predict(X_test) + 1
results["XGBoost"] = accuracy_score(y_test + 1, y_pred)

# LightGBM
y_pred = lgbm.predict(X_test) + 1
results["LightGBM"] = accuracy_score(y_test + 1, y_pred)

# Compare
best_model_name = max(results, key=results.get)
best_accuracy = results[best_model_name]

print("\n=== Model Comparison ===")
for k, v in results.items():
    print(f"{k:20s}: {v:.4f}")

print(f"\nBest Model: {best_model_name} with accuracy {best_accuracy:.4f}")

# Save best model
if best_model_name == "Logistic Regression":
    best_model = logreg
elif best_model_name == "Random Forest":
    best_model = rf
elif best_model_name == "XGBoost":
    best_model = xgb_model
else:
    best_model = lgbm
```

### Step 6 : Predicting Model

```python
train_values = pd.read_csv("train_values.csv")
test_values = train_values.sample(10, random_state=42).copy()  # just 10 samples for demo
test_values["building_id"] = range(1, len(test_values) + 1)

# Preprocess same as training
for col in test_values.columns:
    if test_values[col].dtype == "object":
        test_values[col] = test_values[col].fillna("missing")
    else:
        test_values[col] = test_values[col].fillna(test_values[col].median())

for col, le in encoders.items():
    if col in test_values.columns:
        test_values[col] = test_values[col].apply(lambda x: x if x in le.classes_ else "missing")
        if "missing" not in le.classes_:
            le.classes_ = np.append(le.classes_, "missing")
        test_values[col] = le.transform(test_values[col].astype(str))

X_test_final = test_values.drop(columns=["building_id"], errors="ignore")

# Predict
preds = best_model.predict(X_test_final) + 1  # shift back to {1,2,3}

# Show sample predictions
print("\n=== Sample Predictions ===")
print(pd.DataFrame({
    "building_id": test_values["building_id"],
    "predicted_damage_grade": preds
}).head(10))
```

## Suggestions for Seismologists

### 1. Building Age and Maintenance
- Older buildings tend to be more vulnerable during earthquakes, so they should be prioritized for retrofitting.  
- Schedule regular structural inspections to identify weak points early.  
- Encourage property owners to upgrade unsafe structures and replace outdated materials.

### 2. Building Height and Floors
- Avoid constructing tall, unreinforced multi-story buildings in high-risk seismic areas.  
- Buildings with more than three floors should follow strict seismic-resistant design standards.  
- In high-risk or rural regions, low-rise and flexible structures are generally safer.

### 3. Construction Materials and Superstructure
- Stone, brick, and mud-mortar buildings are more prone to collapse unless reinforced. Techniques like seismic bands, tie beams, and corner strengthening should be applied.  
- Promote high-quality reinforced concrete construction with proper supervision.  
- Weak roofing materials such as thatch or thin metal sheets should be restricted in earthquake-prone zones.

### 4. Soil and Foundation
- Conduct a geotechnical survey before construction begins.  
- Avoid building on loose soil, soft clay, or filled land, as these amplify seismic shaking.  
- Use deeper, well-reinforced foundations in areas likely to experience strong earthquakes.

### 5. Geographic and Regional Planning
- Use geo-level data to identify regions that are historically more prone to seismic damage.  
- Develop detailed seismic microzonation maps to guide future planning decisions.  
- Relocate or retrofit essential facilities in high-hazard zones.

### 6. Critical Infrastructure (Schools, Hospitals, Government Offices)
- Apply stricter seismic design requirements to essential buildings.  
- Carry out rapid visual screening of schools and hospitals to identify vulnerable structures.  
- Make earthquake-resilience audits mandatory before issuing occupancy certificates.

### 7. Policy, Training, and Enforcement
- Strengthen enforcement of building codes, as many failures occur due to poor compliance.  
- Train masons, contractors, and engineers in earthquake-resistant construction techniques.  
- Conduct public awareness campaigns promoting safe construction and retrofitting.

---

**In short: strengthen old structures, ensure new buildings meet safety standards, reinforce weak materials, and use regional data to plan smarter.**
